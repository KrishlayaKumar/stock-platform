```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import os

# -------- USER INPUT --------
stock_symbol = input("Enter stock symbol (e.g., AAPL, RELIANCE.NS): ")

# -------- CREATE DATA FOLDER IF NOT EXISTS --------
save_dir = "data/raw"
os.makedirs(save_dir, exist_ok=True)

# -------- FETCH DATA --------
stock = yf.Ticker(stock_symbol)
data = stock.history(period="5y")

# -------- CHECK IF DATA EXISTS --------
if data.empty:
    print("Invalid stock symbol or no data found.")
    exit()

# -------- SAVE DATA --------
file_path = os.path.join(save_dir, f"{stock_symbol}_data.csv")
data.to_csv(file_path)

print(f" Data saved at: {file_path}")

# -------- DATA PREVIEW --------
print("\n Data Preview:")
print(data.head())

# -------- PLOT CLOSING PRICE --------
plt.figure(figsize=(10,5))
plt.plot(data.index, data["Close"])
plt.title(f"{stock_symbol} Closing Price (5 Years)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import os

# -------- LIST OF STOCK SYMBOLS --------
stocks = [
    "RELIANCE.NS",
    "TCS.NS",
    "INFY.NS",
    "HDFCBANK.NS",
    "ICICIBANK.NS",
    "AAPL",
    "MSFT",
    "GOOGL",
    "AMZN",
    "TSLA"
]

# -------- CREATE DATA DIRECTORY --------
save_dir = "data/raw"
os.makedirs(save_dir, exist_ok=True)

print(" Starting stock data fetch...\n")

for symbol in stocks:
    print(f"Fetching data for {symbol}...")

    stock = yf.Ticker(symbol)
    data = stock.history(period="5y")

    if data.empty:
        print(f" No data found for {symbol}\n")
        continue

    # -------- SAVE CSV --------
    file_path = os.path.join(save_dir, f"{symbol}_data.csv")
    data.to_csv(file_path)

    print(f" Saved: {file_path}")

    # -------- OPTIONAL: PLOT --------
    plt.figure(figsize=(8,4))
    plt.plot(data.index, data["Close"])
    plt.title(f"{symbol} Closing Price (5 Years)")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.grid()
    plt.show()

print("\n All stock data fetching completed.")
```
