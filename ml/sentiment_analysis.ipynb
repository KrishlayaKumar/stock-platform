```python
import yfinance as yf
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# -------- USER INPUT --------
stock_symbol = input("Enter stock symbol (e.g., AAPL, RELIANCE.NS): ")

# -------- FETCH NEWS --------
ticker = yf.Ticker(stock_symbol)
news = ticker.news

if not news:
    print("âŒ No news found for this stock.")
    exit()

# -------- SENTIMENT ANALYZER --------
sia = SentimentIntensityAnalyzer()

sentiments = []

print("\nğŸ“° News Headlines & Sentiment\n")

for article in news[:10]:  # last 10 headlines
    headline = article.get("title", "")
    score = sia.polarity_scores(headline)["compound"]
    sentiments.append(score)

    label = (
        "Positive ğŸ™‚" if score > 0.05 else
        "Negative ğŸ™" if score < -0.05 else
        "Neutral ğŸ˜"
    )

    print(f"- {headline}")
    print(f"  Sentiment: {label} ({score:.2f})\n")

# -------- FINAL SENTIMENT --------
avg_sentiment = sum(sentiments) / len(sentiments)

if avg_sentiment > 0.05:
    final_sentiment = "POSITIVE ğŸ“ˆ"
elif avg_sentiment < -0.05:
    final_sentiment = "NEGATIVE ğŸ“‰"
else:
    final_sentiment = "NEUTRAL âš–ï¸"

print("ğŸ“Š Overall News Sentiment")
print("-------------------------")
print(f"Stock: {stock_symbol}")
print(f"Final Sentiment: {final_sentiment}")
print(f"Average Score: {avg_sentiment:.2f}")
```

```python
import yfinance as yf
from nltk.sentiment.vader import SentimentIntensityAnalyzer

stock_symbol = input("Enter stock symbol (e.g., AAPL, RELIANCE.NS): ")

ticker = yf.Ticker(stock_symbol)
news = ticker.news

sia = SentimentIntensityAnalyzer()

sentiments = []

print("\nNews Headlines & Sentiment\n")

for article in news[:15]:
    headline = article.get("title")

    if not headline or headline.strip() == "":
        continue

    score = sia.polarity_scores(headline)["compound"]
    sentiments.append(score)

    label = (
        "Positive " if score > 0.05 else
        "Negative " if score < -0.05 else
        "Neutral "
    )

    print(f"- {headline}")
    print(f"  Sentiment: {label} ({score:.2f})\n")

# SAFE AGGREGATIO
if len(sentiments) == 0:
    final_sentiment = "NEUTRAL "
    avg_sentiment = 0.0
else:
    avg_sentiment = sum(sentiments) / len(sentiments)

    if avg_sentiment > 0.05:
        final_sentiment = "POSITIVE "
    elif avg_sentiment < -0.05:
        final_sentiment = "NEGATIVE "
    else:
        final_sentiment = "NEUTRAL "

print("Overall News Sentiment")
print("-------------------------")
print(f"Stock: {stock_symbol}")
print(f"Final Sentiment: {final_sentiment}")
print(f"Average Score: {avg_sentiment:.2f}")
```
